```python
tables = [table22, table21, table20, table19, table18, table17, table16, table15]
merged_table = pd.concat(tables, ignore_index=True)


display(merged_table)
```

```python
for col in merged_table.columns:
    merged_table[col] = merged_table[col].map(lambda x: str(x).replace('\n', ' ').replace('\r', ' ').replace('\t', ' ') if pd.notnull(x) else x)

merged_table = merged_table[merged_table['State'] != 'State Name']
merged_table = merged_table[merged_table['Station Code'].notna()]

display(merged_table)



```

```python
merged_table['Station Code'] = merged_table['Station Code'].astype(int)

merged_table['Year'] = merged_table['Year'].astype(int)
```

```python
import pandas as pd

# Example: Convert several columns to float
# Define a dictionary where keys are column names and values are the desired data types
conversion_dict = {
    'Temperature min': float,
    'Temperature max': float,
    'pH min': float,
    'pH max': float,
    'Conductivity min': float,
    'Conductivity max': float,
    'BOD min': float,
    'BOD max': float,
    'Nitrate + Nitrite min': float,
    'Nitrate + Nitrite max': float,
    'Fecal Coliform min': float,
    'Fecal Coliform max': float,
    'Total Coliform min': float,
    'Total Coliform max': float
}

# Apply the conversions
for col, dtype in conversion_dict.items():
    if dtype == float:
        # Use pd.to_numeric for robust conversion to numbers, coercing errors to NaN
        merged_table[col] = pd.to_numeric(merged_table[col], errors='coerce')
    else:
        # For other types, use astype()
        merged_table[col] = merged_table[col].astype(dtype)
```

```python
# Get the total number of unique years in the dataset
num_unique_years = merged_table['Year'].nunique()

# Group by 'Station Code' and count the number of unique years for each station
station_year_counts = merged_table.groupby('Station Code')['Year'].nunique()

# Identify 'Station Code' values that appear in all unique years
common_station_codes = station_year_counts[station_year_counts == num_unique_years].index

# Filter the merged_table to keep only these common station codes
filtered_merged_table = merged_table[merged_table['Station Code'].isin(common_station_codes)]

# Display the first few rows of the filtered table and its shape to confirm
display(filtered_merged_table.head())
print(f"Original table shape: {merged_table.shape}")
print(f"Filtered table shape: {filtered_merged_table.shape}")
```
